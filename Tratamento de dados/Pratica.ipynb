```python
import pandas as pd
import statistics   as sts
import seaborn as srn
```

```python
"""
Iniciando análise exploratória de dados
    Objetivo: Identificar os possíveis problemas a serem tratados na base de dados
"""

"""
Comando "dataset.columns.tolist()" importante para entendermos como as colunas estão nomeadas, exibindo caracteres ocultos se tiver
"""

dataset = pd.read_csv("movies.csv")
dataset.columns.tolist()
```

```python
"""Verificando Tamanho do dataset"""
print(dataset.shape)
print(dataset.size)
```

```python
"""Verificando o datasete em si para identificarmos o conteudo de cada coluna"""
dataset.head()
```

```python
"""
Nessa fase conseguimos identificar algumas inconsistencias de cara, 
como dados NAN e inconsistencias no registro da coluna YEAR


Vamos começar a verificação primeiramente de dados categoricos:

 'MOVIES',
 'YEAR',
 'GENRE',
 'ONE-LINE',
 'STARS',
 
"""

"""MOVIES"""
grup_mv = dataset.groupby(["MOVIES"]).size()
grup_mv

```

```python
grup_mv.sort_values(ascending=False).head(50).plot.bar(color="green")
```

```python
"YEAR"
grup_ye = dataset.groupby(["YEAR"]).size()
grup_ye
```

```python
grup_ye.sort_values(ascending=False).head(40).plot.bar(color="green")
```

```python
"GENRE"
grup_ge = dataset.groupby(["GENRE"]).size()
grup_ge
```

```python
grup_ge.sort_values(ascending=False).head(50).plot.bar(color="green")
```

```python
'ONE-LINE'
grup_on = dataset.groupby(['ONE-LINE']).size()
grup_on
```

```python
grup_on.sort_values(ascending=False).head(5).plot.bar(color="blue")
```

```python
'STARS'
grup_st = dataset.groupby(["STARS"]).size()
grup_st 
```

```python
grup_st.sort_values(ascending=False).head(5).plot.bar(color="blue")
```
