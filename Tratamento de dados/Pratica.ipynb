```python
"""Tratamento de Strings"""
dataset.head(3)
```

```python
"""Removendo todos os caracteres não númericos das colunas numericas e espaçoes em branco se tiver"""
dataset[["YEAR", "RATING", "VOTES", "RunTime", "Gross"]] = dataset[["YEAR", "RATING", "VOTES", "RunTime", "Gross"]].replace(regex=[r"\D|\s"],value="")
```

```python
dataset[["MOVIES","GENRE","ONE-LINE","STARS"]] = dataset[["MOVIES","GENRE","ONE-LINE","STARS"]].replace(regex=[r"\s|\W"],value="")
```

```python
dataset[["MOVIES","GENRE","ONE-LINE","STARS"]] = dataset[["MOVIES","GENRE","ONE-LINE","STARS"]].astype(str)
```

```python
dataset["MOVIES"] = dataset["MOVIES"].str.upper()
dataset["GENRE"]=dataset["GENRE"].str.upper()
dataset["ONE-LINE"]=dataset["ONE-LINE"].str.upper()
dataset["STARS"]=dataset["STARS"].str.upper()
```

```python
s = dataset.loc[dataset["YEAR"].str.len() > 4,"YEAR"]

dataset.loc[dataset["YEAR"].str.len() > 4, "YEAR"]= dataset.loc[dataset["YEAR"].str.len() > 4, "YEAR"].str[:-4]
```

```python
"""Iniciando tratamento de duplicatas"""
dataset[dataset.duplicated(['MOVIES',
 'YEAR',
 'GENRE',
 'RATING',
 'ONE-LINE',
 'STARS',
 'VOTES',
 'RunTime',
 'Gross'],keep=False)]
```

```python
dataset.drop_duplicates(subset=
['MOVIES',
 'YEAR',
 'GENRE',
 'RATING',
 'ONE-LINE',
 'STARS',
 'VOTES',
 'RunTime',
 'Gross'],
 keep="first",inplace=True)
```
