```python
import numpy as np
import pandas as pd
import statistics as sts
import seaborn as srn
import matplotlib.pyplot as plt
import plotly.express as px
import nbformat
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
df = pd.read_csv(r"VENDAS2.csv")
df['DATA'] = pd.to_datetime(df['DATA'])
df.head()
```

```python
df
```

```python
df_mensal = df.groupby(df['DATA'].dt.to_period('M'))['TOTAL'].sum().reset_index()
df_mensal.columns
```

```python
df_mensal = df.groupby(df['DATA'].dt.to_period('M'))['TOTAL'].sum().reset_index()
df_mensal.columns = ["Periodos", "Fechamento"]
df_mensal['Periodos'] = df_mensal['Periodos'].dt.to_timestamp()
df_mensal['Periodos'] = df_mensal['Periodos'].dt.strftime(r'%b/%Y')
figure = px.bar(df_mensal,x="Periodos",y="Fechamento", title="Fechamento todo periodo")
figure.show()


```
