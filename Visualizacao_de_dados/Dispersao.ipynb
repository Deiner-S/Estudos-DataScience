```python
"""
PARA QUE SERVE
O gráfico de dispersão (ou scatter plot) serve pra mostrar visualmente a relação entre duas variáveis numéricas. 
Ele responde perguntas do tipo:
    --> Será que quando uma variável aumenta, a outra também aumenta?
    --> Ou será que quando uma aumenta, a outra diminui?
    --> Ou... será que não há relação nenhuma?

    
COMO ANÁLIZAR O GRAFICO:
    --> Se os pontos descem da esquerda pra direita, a relação é negativa
    --> Se os pontos sobem, a relação é positiva.
    --> Se os pontos estão todos bagunçados e sem padrão... pode ser que não tenha relação nenhuma.
"""
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("trees.csv")
df.head()
```

```python
plt.scatter(
    x=df.Girth, 
    y=df.Volume,
    edgecolor='black',
    alpha=0.8,s=100
    )
plt.title("Arvore")
plt.xlabel("Circunferencia")
plt.ylabel("Volume")
```

```python
plt.plot(df.Girth,df.Volume)
plt.title("Arvore")
plt.xlabel("Circunferencia")
plt.ylabel("Volume")
```

```python

# Gráfico de dispersão com 3 colunas
plt.figure(figsize=(10,6))         #regula as dimens"oes da exibição do gráfico
scatter = plt.scatter(
    x=df.Girth,                      # Eixo X
    y=df.Volume,                       # Eixo Y
    c=df.Height,                    # Cor dos pontos
    cmap='coolwarm',                  # Mapa de cores
    s=100,                           # Tamanho dos pontos
    edgecolor='black',
    alpha=0.8
)

# Personalizando o gráfico
plt.colorbar(scatter, label='Altura')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
plt.title('Relação entre Circunferencia, Altura e volume das arvores')
plt.grid(True)
plt.tight_layout()
plt.show()
```
